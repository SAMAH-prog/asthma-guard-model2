```python
import os

# 1. Écriture du fichier de configuration
with open("config.py", "w") as f:
    f.write('''# Configuration des variables et plages réalistes
N_PATIENTS = 5
HOURS_PER_PATIENT = 720  # ~30 jours de données horaires
RANDOM_SEED = 42

# Plages physiologiques et environnementales conformes au document
FC_REPOS_MEAN = 75
FC_REPOS_STD = 15
SPO2_MEAN = 96
SPO2_STD = 2
PM25_MIN = 0
PM25_MAX = 150
''')

# 2. Écriture du module dataset
with open("dataset.py", "w") as f:
    f.write('''import numpy as np
import pandas as pd
from config import *

def generate_patient_series(patient_id):
    np.random.seed(RANDOM_SEED + patient_id)

    # Plage temporelle horaire
    date_range = pd.date_range(start="2026-05-01", periods=HOURS_PER_PATIENT, freq="h")

    # Signaux physiologiques (Lois normales et distributions réalistes)
    fc_repos = np.random.normal(FC_REPOS_MEAN, FC_REPOS_STD, HOURS_PER_PATIENT).clip(50, 120)
    spo2 = np.random.normal(SPO2_MEAN, SPO2_STD, HOURS_PER_PATIENT).clip(85, 100)
    activite = np.random.negative_binomial(n=5, p=0.01, size=HOURS_PER_PATIENT).clip(0, 5000)
    pm25 = np.random.gamma(shape=2, scale=20, size=HOURS_PER_PATIENT).clip(PM25_MIN, PM25_MAX)

    df = pd.DataFrame({
        "timestamp": date_range,
        "patient_id": patient_id,
        "fc_repos": fc_repos,
        "spo2": spo2,
        "activite": activite,
        "pm25": pm25,
        "heure": date_range.hour
    })

    # Logique causale : Risque accru selon les seuils critiques [cite: 127]
    score_risque = (
        0.3 * (df["fc_repos"] > 95) +
        0.4 * (df["spo2"] < 93) +
        0.3 * (df["pm25"] > 80)
    )

    # Effet causal décalé dans le temps (horizon de 4 à 24 heures) [cite: 25]
    df["crise_imminente"] = (score_risque.shift(-12, fill_value=0) + np.random.normal(0, 0.1, HOURS_PER_PATIENT) > 0.4).astype(int)

    return df

def get_multivariate_dataset():
    all_patients = [generate_patient_series(i) for i in range(N_PATIENTS)]
    return pd.concat(all_patients, ignore_index=True)
''')

print("Modules 'config.py' et 'dataset.py' réinitialisés sans erreur.")
```

```python
with open("model.py", "w") as f:
    f.write('''import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from xgboost import XGBClassifier
from dataset import get_multivariate_dataset

def train_and_evaluate_model():
    df = get_multivariate_dataset()

    # Isolation du Patient 0 pour l'évaluation
    df_p0 = df[df["patient_id"] == 0].copy()
    split_idx = int(len(df_p0) * 0.8)

    train_df = df_p0.iloc[:split_idx]
    test_df = df_p0.iloc[split_idx:]

    features = ["fc_repos", "spo2", "activite", "pm25", "heure"]

    X_train, y_train = train_df[features], train_df["crise_imminente"]
    X_test, y_test = test_df[features], test_df["crise_imminente"]

    # Entraînement du classificateur
    clf = XGBClassifier(n_estimators=100, max_depth=5, random_state=42, eval_metric="logloss")
    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_test)

    metrics = {
        "accuracy": accuracy_score(y_test, y_pred),
        "f1": f1_score(y_test, y_pred),
        "report": classification_report(y_test, y_pred, output_dict=True),
        "matrix": confusion_matrix(y_test, y_pred),
        "importance": clf.feature_importances_
    }

    return metrics, features, y_test.values, y_pred
''')

print("Module 'model.py' mis à jour.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from model import train_and_evaluate_model

# Lancement de l'évaluation
metrics, features, y_true, y_pred = train_and_evaluate_model()

# 1. Rapport textuel des performances
print("=== PERFORMANCES DU MODÈLE 2 (ASTHMAGUARD) ===")
print(f"Précision globale (Accuracy) : {metrics['accuracy']:.4f}")
print(f"F1-Score : {metrics['f1']:.4f}\n")
print("Détails par classe :")
for label, item in metrics['report'].items():
    if isinstance(item, dict):
        print(f" Classe {label} -> Précision: {item['precision']:.2f}, Rappel: {item['recall']:.2f}")

# 2. Affichage des diagrammes d'analyse
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Matrice de confusion
sns.heatmap(metrics['matrix'], annot=True, fmt='d', cmap='Blues', ax=ax[0],
            xticklabels=['Pas de crise', 'Crise Imminente'],
            yticklabels=['Pas de crise', 'Crise Imminente'])
ax[0].set_title("Matrice de Confusion")
ax[0].set_xlabel("Prédictions du modèle")
ax[0].set_ylabel("Réalité terrain")

# Importance des variables
ax[1].barh(features, metrics['importance'], color='indigo')
ax[1].set_title("Importance des variables (Signaux Causaux)")
ax[1].set_xlabel("Poids explicatif")

plt.tight_layout()
plt.show()
```

```python
import asyncio
import nest_asyncio
import numpy as np
import uvicorn
from fastapi import FastAPI
from pydantic import BaseModel

# 1. Application indispensable du patch asynchrone pour Jupyter/Colab
nest_asyncio.apply()

app = FastAPI(title="AsthmaGuard API Temps Réel")


class WatchDataInput(BaseModel):
    fc_repos: float
    spo2: float
    activite: float
    pm25: float
    heure: int


@app.post("/predict")
def predict_crisis(data: WatchDataInput):
    score = 0.0

    # Nettoyage strict des conditions métier (Modèle 2)
    if data.fc_repos > 95:
        score += 0.35
    if data.spo2 < 93:
        score += 0.45
    if data.pm25 > 80:
        score += 0.20

    prob = min(max(score + np.random.normal(0, 0.02), 0.0), 1.0)

    # Grille d'évaluation des alertes AsthmaGuard
    if prob >= 0.70:
        statut = "Rouge (Alerte Risque Très Élevé)"
    elif prob >= 0.45:
        statut = "Orange (Surveillance Renforcée)"
    elif prob >= 0.25:
        statut = "Jaune (Prudence)"
    else:
        statut = "Vert (Pas d'alerte)"

    return {
        "status": "success",
        "crisis_probability": round(prob, 2),
        "urgency_level": statut,
        "time_horizon": "4-24 heures",
    }


# 2. Lancement sécurisé d'Uvicorn dans la boucle active de Colab
if __name__ == "__main__":
    print("Serveur FastAPI AsthmaGuard actif sur http://127.0.0.1:8000")

    config = uvicorn.Config(app=app, host="127.0.0.1", port=8000, loop="asyncio")
    server = uvicorn.Server(config)

    # Récupération de la boucle de Colab pour y greffer le serveur sans blocage
    loop = asyncio.get_event_loop()
    loop.create_task(server.serve())
```
