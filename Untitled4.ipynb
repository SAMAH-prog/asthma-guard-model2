```python
import os

with open("config.py", "w") as f:
    f.write('''# Configuration
N_PATIENTS = 5
HOURS_PER_PATIENT = 1000  # Plus de données pour stabiliser les séquences
RANDOM_SEED = 42

FC_REPOS_MEAN = 75
FC_REPOS_STD = 12
SPO2_MEAN = 95.5
SPO2_STD = 2
PM25_MIN = 10
PM25_MAX = 150
''')

with open("dataset.py", "w") as f:
    f.write('''import numpy as np
import pandas as pd
from config import *

def generate_patient_series(patient_id):
    np.random.seed(RANDOM_SEED + patient_id)
    date_range = pd.date_range(start="2026-05-01", periods=HOURS_PER_PATIENT, freq="h")

    fc_repos = np.random.normal(FC_REPOS_MEAN, FC_REPOS_STD, HOURS_PER_PATIENT).clip(50, 120)
    spo2 = np.random.normal(SPO2_MEAN, SPO2_STD, HOURS_PER_PATIENT).clip(85, 100)
    activite = np.random.negative_binomial(n=5, p=0.01, size=HOURS_PER_PATIENT).clip(0, 2000)
    pm25 = np.random.gamma(shape=2, scale=25, size=HOURS_PER_PATIENT).clip(PM25_MIN, PM25_MAX)

    df = pd.DataFrame({
        "timestamp": date_range,
        "patient_id": patient_id,
        "fc_repos": fc_repos,
        "spo2": spo2,
        "activite": activite,
        "pm25": pm25,
        "heure": date_range.hour
    })

    # Effet cumulatif et séquentiel marqué sur les 6 dernières heures
    pm25_accumule = df["pm25"].rolling(window=6, min_periods=1).mean()
    spo2_historique_bas = df["spo2"].rolling(window=6, min_periods=1).mean()

    score_causal = (
        0.50 * (spo2_historique_bas < 95.0) +
        0.35 * (pm25_accumule > 50) +
        0.15 * (df["fc_repos"] > 85)
    )

    # Génération d'environ 15-20% de cas de crise à l'horizon +12h
    df["crise_imminente"] = (score_causal.shift(-12, fill_value=0) > 0.42).astype(int)

    return df

def get_multivariate_dataset():
    all_patients = [generate_patient_series(i) for i in range(N_PATIENTS)]
    return pd.concat(all_patients, ignore_index=True)
''')

print("Fichier dataset.py corrigé pour un meilleur équilibre des classes.")
```

```python
with open("model.py", "w") as f:
    f.write('''import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from dataset import get_multivariate_dataset

def train_and_compare_models():
    df = get_multivariate_dataset()
    df_p0 = df[df["patient_id"] == 0].copy()

    features_brutes = ["fc_repos", "spo2", "activite", "pm25", "heure"]

    # Reconstruction de l'historique séquentiel (Lags temporels type TFT)
    df_causal = df_p0.copy()
    features_causales = list(features_brutes)
    for lag in [1, 2, 3, 4, 6]:
        for col in ["spo2", "pm25", "fc_repos"]:
            df_causal[f"{col}_lag_{lag}"] = df_causal[col].shift(lag)
            features_causales.append(f"{col}_lag_{lag}")

    df_causal = df_causal.dropna()

    split_idx = int(len(df_causal) * 0.8)

    X_train_c = df_causal[features_causales].iloc[:split_idx]
    y_train_c = df_causal["crise_imminente"].iloc[:split_idx]
    X_test_c = df_causal[features_causales].iloc[split_idx:]
    y_test_c = df_causal["crise_imminente"].iloc[split_idx:]

    X_train_rf = df_causal[features_brutes].iloc[:split_idx]
    X_test_rf = df_causal[features_brutes].iloc[split_idx:]

    # Calcul du facteur d'échelle pour équilibrer la classe minoritaire dans XGBoost
    nb_neg = np.sum(y_train_c == 0)
    nb_pos = np.sum(y_train_c == 1)
    poids_echelle = nb_neg / max(1, nb_pos)

    # 1. Modèle Temporel Causal (XGBoost fenêtré + Équilibrage)
    clf_causal = XGBClassifier(
        n_estimators=150,
        max_depth=5,
        learning_rate=0.05,
        scale_pos_weight=poids_echelle,
        random_state=42,
        eval_metric="logloss"
    )
    clf_causal.fit(X_train_c, y_train_c)
    y_pred_causal = clf_causal.predict(X_test_c)

    # 2. Random Forest Standard (Instant T seul + Équilibrage)
    clf_rf = RandomForestClassifier(
        n_estimators=150,
        max_depth=5,
        class_weight="balanced",
        random_state=42
    )
    clf_rf.fit(X_train_rf, y_train_c)
    y_pred_rf = clf_rf.predict(X_test_rf)

    return {
        "main": {
            "name": "Modèle Temporel Causal (TFT)",
            "accuracy": accuracy_score(y_test_c, y_pred_causal),
            "precision": precision_score(y_test_c, y_pred_causal, zero_division=0),
            "recall": recall_score(y_test_c, y_pred_causal, zero_division=0),
            "f1": f1_score(y_test_c, y_pred_causal, zero_division=0),
            "matrix": confusion_matrix(y_test_c, y_pred_causal)
        },
        "rf": {
            "name": "Random Forest (Standard)",
            "accuracy": accuracy_score(y_test_c, y_pred_rf),
            "precision": precision_score(y_test_c, y_pred_rf, zero_division=0),
            "recall": recall_score(y_test_c, y_pred_rf, zero_division=0),
            "f1": f1_score(y_test_c, y_pred_rf, zero_division=0),
            "matrix": confusion_matrix(y_test_c, y_pred_rf)
        }
    }
''')

print("Fichier model.py mis à jour avec correction du déséquilibre des classes.")
```
